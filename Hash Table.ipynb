<h1><center>Hash Table</h1></center>

```python
class HashTable:
    def __init__(self, size = 7):
        self.data_map = [None] * size
        # create 7 lists with nothing
        
    def __hash(self, key):
        my_hash = 0
        for letter in key:
            my_hash = (my_hash + ord(letter) * 23) % len(self.data_map)
            # Using 23 because 23 is the prime number, can use any other prime number
        return my_hash
    
    def print_table(self):
        for i, val in enumerate(self.data_map):
            print(i,": ", val)
            
    def set_item(self,key, value):
        index = self.__hash(key) # apply hash function
        if self.data_map[index] == None:
            self.data_map[index] = []
            # create a empty list if there is no list
            # Ex: 0: None means there is no list
            
        self.data_map[index].append([key,value])
        # Passing the key and value in a list to the index list
        # linear separating
        
        
    def get_item(self,key):
        index = self.__hash(key)
        if self.data_map[index] is not None:
            for i in range(len(self.data_map[index])):
                if self.data_map[index][i][0] == key:
                    return self.data_map[index][i][1] #[1] target for the value
        return None # if target is not in the table
    
    def keys(self): # return all the keys in a litst
        all_keys = []
        for i in range(len(self.data_map)):
            if self.data_map[i] is not None:
                for j in range(len(self.data_map[i])):
                    all_keys.append(self.data_map[i][j][0])
        return all_keys
                
```

```python
mh = HashTable()
```

```python
mh.print_table()
```

```python
mh.set_item("bolts", 1400)
mh.set_item("washers", 50)
mh.set_item("lumber", 70)
```

```python
mh.print_table()
```

```python
print(mh.get_item("bolts"))
print(mh.get_item("washers"))
print(mh.get_item("lumber"))
print(mh.get_item("car"))
```

# Interview Question

Find the same value in two lists

## Method 1: Nested For Loop

```python
# return only one value
def nested_for_loop(list1, list2):
    for i in list1:
        for j in list2:
            if i == j:
                return i

list1 = [1,3,5]
list2 = [2,4,5]
```

```python
print(nested_for_loop(list1, list2))
```

```python
# return only all value that are in both lists
def nested_for_loop_2(list1, list2):
    result = []
    for i in list1:
        for j in list2:
            if i == j:
                result.append(i)
    return result

list1 = [1,3,5,9,0,2]
list2 = [2,4,5,0,9,11]
```

```python
print(nested_for_loop_2(list1, list2))
```

## Method 2: Hash Table

```python
def hash_for_common_item(list1, list2):
    d = {}
    result = []
    for i in list1:
        d[i] = True
    for j in list2:
        if j in d:
            result.append(j)
    return result

print(hash_for_common_item(list1,list2))
```

Now it is not a nested loop, so the Big O is 2n which can be considerd as O(n)
