```python
# Set up

class Node:
    def __init__(self,value):
        self.value = value
        self.left = None
        self.right = None
        # similar to the node for Double Linked List

        
class BinarySearchTree:
    
    """
    This method add the first node as initializing the function
    def __init__(self, value):
        new_node = Node(value)
        self.root = new_node # set the first element as the root
    """
    def __init__(self):
        self.root = None
        # This method will add the root and value later and create an empty tree
        
    def insert(self, value):
        new_node = Node(value)
        
        # situation when empty tree
        if self.root == None:
            self.root = new_node
            return True
        
        temp = self.root
        while (True):
            if new_node.value == temp.value: # Check for duplicate value
                return False
            if new_node.value < temp.value:
                if temp.left is None: # If there is no node under the spot
                    temp.left = new_node
                    return True
                temp = temp.left # If there is a node, we move on to the next level (still left side)
            else:
                if temp.right is None: # check available spot
                    temp.right = new_node
                    return True
                temp = temp.right # move to next level
                
    def contains(self, value): # find if the value is in the tree or not
        
        temp = self.root # pointer, also if the tree is empty then the temp would be None, and skip the while loop
        while temp is not None:
            if value < temp.value: # If value is less than the target we go left
                temp = temp.left
            elif value > temp.value:
                temp = temp.right # larger and go right
            else: # equal to the value
                return True
        return False # if the tree is empty it will skip the while loop and return False here
```

```python
tree = BinarySearchTree()
print(tree.root)
```

```python
tree.insert(2)
tree.insert(1)
tree.insert(3)
```

```python
print(tree.root.value)
```

```python
print(tree.root.right.value)
```

```python
print(tree.root.left.value)
```
